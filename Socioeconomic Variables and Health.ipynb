## Step 3: Visualising the Data

Part 1) Here we will create a heatmap to show the correlation between the 3 socioeconomic variables we have chosen, and health status as a whole. The method to calculate 'BSh' using several indicators is outlined in the methdology section of our website.

The numerical value in the centre of each square is the Pearson Correlation Co-efficient. The further away the value is from 0, the stronger relationship between the two variables. A positive value shows a positive correlation, whilst a negative value indicates a negative correlation

Source for the Heatmap Guide: Waskom, a, 2022. seaborn.heatmap â€” seaborn 0.11.2 documentation. [online] Seaborn.pydata.org. Available at: https://seaborn.pydata.org/generated/seaborn.heatmap.html [Accessed 4 January 2022].

```python
corr = df.corr(method='pearson')
plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True)
plt.yticks(rotation=0) 
```

Part 2) Now we will create a pairplot to visualise the relationships between these variables, the correlations for which are seen in the heatmap above. The diagonals in the pairplot show histograms to visualise the distribution of each variable's data.

Source for the Pairplot Guide: Waskom, b, 2022. seaborn.pairplot â€” seaborn 0.11.2 documentation. [online] Seaborn.pydata.org. Available at: https://seaborn.pydata.org/generated/seaborn.pairplot.html [Accessed 6 January 2022].

```python
sns.set_palette("Spectral", 12)
sns.pairplot(data, vars=["Income", "Education", "Race & Migration", "BSh"], kind="reg")
plt.show()
```
