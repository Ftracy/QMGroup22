```python
import pandas
pandas.__version__
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Downloads/Life expectancy at birth by London borough (2017-19).csv")
data.head()
```

## Step 2: Cleaning the Data and Creating a Data Frame

```python
data= data.drop(["Life expectancy at birth by borough for men and women (2017-19)"],axis=1)
```

```python
data= data.drop([0])
```

```python
new_header = data.iloc[0]
data = data[1:] 
data.columns = new_header
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
print (df)
print (df.dtypes)
```

#### ^^ Here it is important to note that the values for Life Expectancy and Healthy Life Expectancy appear as 'objects'. In order to plot/graph the data, these values need to be recognised as numerical

```python
df = pd.DataFrame(data)
df['Life Expectancy (LE)'] = pd.to_numeric(df['Life Expectancy (LE)'])
df['Healthy Life Expectancy (HLE)'] = pd.to_numeric(df['Healthy Life Expectancy (HLE)'])

print (df)
print (df.dtypes)
```

#### Now the values for Life Expectancy and Healthy Life Expectancy are recognised as 'floats'

## Step 3: Visualising the Distribution of this Data

```python
ax = sns.boxplot(y="Sex", x="Life Expectancy (LE)", data=data, palette="PuBuGn")
ax = sns.swarmplot(y="Sex", x="Life Expectancy (LE)", data=data, color=".6")
plt.title("Boxplot Overlaid with a Swarmplot showing Life Expectancy (LE) Across London Boroughs")
```

```python
ax = sns.boxplot(y="Sex", x="Healthy Life Expectancy (HLE)", data=data, palette="PuBuGn")
ax = sns.swarmplot(y="Sex", x="Healthy Life Expectancy (HLE)", data=data, color=".6")
plt.title("Boxplot Overlaid with a Swarmplot to show Healthy Life Expectancy (HLE) Across London Boroughs")
```

```python
ax = sns.boxplot(x="Life Expectancy (LE)", data=data, palette="PuBuGn")
ax = sns.stripplot(x="Life Expectancy (LE)", data=data, color=".3")
```

```python
ax = sns.boxplot(x="Healthy Life Expectancy (HLE)", data=data, palette="PuBuGn")
ax = sns.stripplot(x="Healthy Life Expectancy (HLE)", data=data, color=".3")
```
