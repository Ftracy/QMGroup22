```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('SOCIO-ECON.csv', encoding='latin1')
print("Size of dataset is : ", df.shape)
df
df_all_boroughs=df[['ï»¿Boroughs', 'Income', 'Education', 'Race&Migration', 'Health']]
df_all_boroughs= df_all_boroughs.rename(columns={'ï»¿Boroughs': 'Boroughs'})
df_all_boroughs
import sklearn
from sklearn import linear_model
X = df_all_boroughs[['Income', 'Education', 'Race&Migration']] 
# here we have 3 variables for multiple regression. 
# being Green Space Score, Air Pollution Score, and Noise Pollution Score
Y = df_all_boroughs['Health']
# health (health scores) are calcularted based on exisiting data via approahces discussed in the data methodology section
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
import statsmodels.api as sm
X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y.astype(float),X.astype(float)).fit()
 
print_model = model.summary()
print(print_model)
#R square, coefficient, t value, P>t
#recording of last workshop, scatter graph of each variable
```
